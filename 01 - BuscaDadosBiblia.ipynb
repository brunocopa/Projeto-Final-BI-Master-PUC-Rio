```python
#Buscar todos os livros da API (66 livros)
response = requests.get('https://www.abibliadigital.com.br/api/books',headers=Token)
livros = response.json()
```

```python
lista_versos = []
for livro in livros:
  abbrev    = livro['abbrev']['pt']
  name      = livro['name']
  group     = livro['group']
  author    = livro['author']
  chapters   = livro['chapters']
  testament = livro['testament'] 
  print(abbrev + ' ' + author + ' ' + str(chapters) + ' ' + testament) 
  

  for chapter in range(1,chapters+1):
    try:
      url = "https://www.abibliadigital.com.br/api/verses/" + version + "/" + abbrev + "/" + str(chapter)
      print(url)
      response = requests.get(url,headers = Token)
      versos = response.json()['verses']
      
      for verso in versos: 
        verso_numero = verso['number']
        verso_texto  = verso['text']
        print(str(chapter) + ' ' + str(verso_numero) + ' '  + verso['text'])
        dados_verso = {'abbrev': abbrev,'name': name, 'group': group ,'author': author,'testament': testament, 'chapter': chapter,'verse': verso_numero,'text': verso_texto}
        lista_versos.append(dados_verso)
    
    except Exception as e:
      if type(e) in EXCEPTIONS:
        print('Pulando: {}'.format(url))
        continue
    
    

  #df_versos = pd.DataFrame.from_dict(lista_versos)
  #name_arq = abbrev + '_' + 'versosBiblia.xls'
  #df_versos.to_excel(workdir_path+'/'+name_arq, index = False, header=True)
  
df_versos = pd.DataFrame.from_dict(lista_versos)
df_versos.head()
df_versos.to_csv('versosBliblicosCompleto.csv',encoding="utf-8", index = False, header=True)

 


```
